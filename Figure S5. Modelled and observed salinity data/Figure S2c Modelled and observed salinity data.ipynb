```python
#packages
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Widening the screen
from matplotlib import rcParams
%matplotlib inline
%config InlineBackend.figure_format = 'svg'
# Widening the screen
from IPython.core.display import display, HTML
display(HTML("<style>.container { width:90% !important; }</style>"))
FS=20
rcParams['figure.dpi'] = 1800
rcParams['savefig.dpi'] = 800
rcParams['font.size'] = FS
rcParams['font.family'] = 'StixGeneral'
rcParams["mathtext.fontset"] = 'stix'
# rcParams["mathtext.default"] = "rm" 
# rcParams['xtick.direction'] = 'in'
plt.rc('font', size=FS)          # controls default text sizes
plt.rc('axes', titlesize=FS)     # fontsize of the axes title
plt.rc('axes', labelsize=FS)     # fontsize of the x and y labels
plt.rc('xtick', labelsize=FS)    # fontsize of the tick labels
plt.rc('ytick', labelsize=FS)    # fontsize of the tick labels
plt.rc('legend', fontsize=FS)    # legend fontsize
plt.rc('figure', titlesize=FS)   # fontsize of the figure title
```

```python
import pandas
x1990 = pandas.read_csv('SalinityBefore1990x.csv',header=0)
y1990 = pandas.read_csv('SalinityBefore1990y.csv',header=0)
x2010 = pandas.read_csv('SalinityAfter2010x.csv',header=0)
y2010 = pandas.read_csv('SalinityAfter2010y.csv',header=0)
SalinitySim2020x = pandas.read_csv('SalinitySim2020x.csv',header=0)
SalinitySim2020y = pandas.read_csv('SalinitySim2020y.csv',header=0)
```

```python
nticker = 6
from matplotlib import ticker
fig, ax = plt.subplots(1, 1, figsize=(12, 6))

###
lenSalinitySim2020x = len(SalinitySim2020x)
inter = 20 # 间隔
SalinitySim2020xinter = SalinitySim2020x[0:lenSalinitySim2020x:inter]
SalinitySim2020yinter = SalinitySim2020y[0:lenSalinitySim2020x:inter]
###

pSalData, = ax.plot(x2010, y2010, 'bo', markerfacecolor='none', markersize=10, label='Data')
# pSalSim,  = ax.plot(SalinitySim2020x, SalinitySim2020y, 'k--', lw=3, label='Simulation') 
pSalSim,  = ax.plot(SalinitySim2020xinter, SalinitySim2020yinter, 'c--', lw=3, marker=">", label='Simulation') 

# import seaborn as sns
# sns.regplot(x=x2010, y=y2010, marker="o", order=4, ci=60, truncate=False)

ax.set_xlim(2.0, 12.0)
ax.set_xticks(np.linspace(2.0, 12.0, nticker))

ax.set_ylim(0.0, 2.5)
ax.set_yticks(np.linspace(0.0, 2.5, nticker))

# ax.set_title('Salinity of data and simulation')
ax.set_xlabel("Distance from land to sea ($km$)")
ax.set_ylabel("Salinity (%)")

ax.spines['bottom'].set_linewidth(1.5);
ax.spines['left'].set_linewidth(1.5);
ax.spines['right'].set_linewidth(1.5);
ax.spines['top'].set_linewidth(1.5);

ax.legend(fontsize=FS,frameon=False,ncol=1,loc=0)
ax.grid()

# plt.show()
plt.savefig('Figure S2c Modelled and observed salinity data.pdf', bbox_inches='tight')
```
